```python
age = df[df['variable']=='Age']
gender = df[df['variable']=='Gender']
Blood_Type = df[df['variable']=='Blood Type']
medical_condition = df[df['variable']=='Medical Condition']
admission = df[df['variable']=='Date of Admission']
docter = df[df['variable']=='Doctor']
hospital = df[df['variable']=='Hospital']
insurance_provider = df[df['variable']=='Insurance Provider']
bill_amount = df[df['variable']=='Billing Amount']
room_num = df[df['variable']=='Room Number']
admission_type = df[df['variable']=='Admission Type']
discharge = df[df['variable']=='Discharge Date']
medication = df[df['variable']=='Medication']
result = df[df['variable']=='Test Results']
ethnicity = df[df['variable']=='Ethnicity']
```

```python
with pd.ExcelWriter("NewDf.xlsx") as writer:
   
    # use to_excel function and specify the sheet_name and index 
    # to store the dataframe in specified sheet
    gender.to_excel(writer, sheet_name="Gender", index=False)
    Blood_Type.to_excel(writer, sheet_name="Blood Type", index=False)
    medical_condition.to_excel(writer, sheet_name="Medical Condition", index=False)
    admission.to_excel(writer, sheet_name="Date of Admission", index=False)
    docter.to_excel(writer, sheet_name="Doctor", index=False)
    hospital.to_excel(writer, sheet_name="Hospital", index=False)
    insurance_provider.to_excel(writer, sheet_name="Insurance Provider", index=False)
    bill_amount.to_excel(writer, sheet_name="Billing Amount", index=False)
    room_num.to_excel(writer, sheet_name="Room Number", index=False)
    admission_type.to_excel(writer, sheet_name="Admission Type", index=False)
    discharge.to_excel(writer, sheet_name="Discharge Date", index=False)
    medication.to_excel(writer, sheet_name="Medication", index=False)
    result.to_excel(writer, sheet_name="Test Results", index=False)
    ethnicity.to_excel(writer, sheet_name="Ethnicity", index=False)
    age.to_excel(writer, sheet_name="Age", index=False)
```
