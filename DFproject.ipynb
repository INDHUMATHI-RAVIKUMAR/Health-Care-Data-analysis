<h2> One sample T-test</h2>

```python
from scipy.stats import ttest_1samp
hm = 51.4  # HYPOTHESISED MEAN of Age category 
t_stat, p_value = ttest_1samp(df['Age'],hm)
alpha = 0.05 # acceptance error percent
print(f"T-statistic:{t_stat}")
print(f"P-value:{p_value}")
if p_value<alpha:
    print(f"Reject null hypothesis at alpha ={alpha}")
else:
    print(f"Faild to reject null hypothesis at alpha = {alpha}")
```

<h3> Since the p value is greater than the significance level 0.05, we fail to reject the null hypothesis</h3>
<h4> If we increase the Significance level the Hypothesis result may be change at mean value</h4>

<h2> Independent T-test</h2>

```python
import numpy as np
import scipy.stats as stats
male= df[df['Gender']=='Male']
female= df[df['Gender']=='Female']
male_region = male['Billing Amount']
female_region = female['Billing Amount']
male_group = np.array(male_region)
female_group = np.array(female_region)
statistic,p_value = stats.ttest_ind(a=male_region, b=female_region, equal_var = True) # independant 2 catogories 
print("P-VALUE",p_value,"STATISTIC-VALUE",statistic)
alpha = 0.05
if p_value < alpha:
    print(f"Reject null hypothesis at alpha ={alpha}")
else:
    print(f"Faild to reject null hypothesis at alpha = {alpha}")
```

<h3> Since the p value is greater than the significance level 0.05, we fail to reject the null hypothesis.</h3>
<h3> This means that, based on the data, there is not enough evidence to conclude that there is a significant difference in the mean billing amounts between males and females </h3>

<h1> CHI-SQUARE TEST</h1>

<h3> The chi-square test is used to test the independence of two categorical variables</h3>

```python
import numpy as np
import scipy.stats as stats
from scipy.stats import chi2_contingency
```

```python
contingency_table = pd.crosstab(df["Hospital"],df["Insurance Provider"])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f'Chi-square stattistic:{chi2}')
print(f'P_value :{p}')
print(f'Degrees of freedom:{dof}')
print('Expected frequencies:')
print(expected)
```

```python
if dof < chi2:
    print(f"Reject null hypothesis at alpha ={alpha}")
else:
    print(f"Faild to reject null hypothesis at alpha = {alpha}")
```

<h2> P value is extremely small which is significantly less than the typical significance level of 0.05.</h2>
<h3>This indicates strong evidence to reject the Null hypothesis.</h3>
<h3>The result suggest that the choice of Hospital and Insurance provider are not independent of each other </h3>
